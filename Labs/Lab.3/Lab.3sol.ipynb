There is plenty of material online that will help you figure out how to do various tasks on the command line. Commands you may need to know today:

* `ls`: lists the contents of the current directory.

Write a function that computes the variance of values in a list. Recall the equation for the variance of a random variable $\bf{x}$ computed on a data set of $n$ values $\{ x_i \} = \{x_1, x_2, ..., x_n\}$  is ${\bf\langle x \rangle} = \frac{1}{n} \sum_i^n (x_i - {\bf\bar{x}})$.

````python
def variance(data):
    m = 0
    # create list
    # lst = [1, 22, 55, 43, 67, 96, 26, 41]
    x = len(data)
    # printing length
    print('The length of list is', x)
    # sum of list
    total = sum(data)
    # mean
    m = total / x
    
    # deviation
    dev = [(z - m) ** 2 for z in data]

    # variance
    var = sum(dev) / x

    return var

# Testing
# data = [1, 22, 55, 43, 67, 96, 26, 41]
# print("Variance of Data:", variance(data))
````

## Histogramming

*Exercise 3:* Write a function that bins the data so that you can create a histogram. An example of how to implement histogramming is the following logic:

* User inputs a list of values `x` and optionally `n_bins` which defaults to 10.
* If not supplied, find the minimum and maximum (`x_min`,`x_max`) of the values in x.
* Determine the bin size (`bin_size`) by dividing the range of the function by the number of bins.
* Create an empty list of zeros of size `n_bins`, call it `hist`.
* Loop over the values in `x`
    * Loop over the values in `hist` with index `i`:
        * If x is between `x_min+i*bin_size` and `x_min+(i+1)*bin_size`, increment `hist[i].` 
        * For efficiency, try to use continue to goto the next bin and data point.
* Return `hist` and the list corresponding of the bin edges (i.e. of `x_min+i*bin_size`).    

````python
#Solution
import numpy as np

def histogram(x, n_bins=10, x_min=None, x_max=None):
    ### BEGIN SOLUTION
    if x_min is None:
        x_min = min(x)
    if x_max is None:
        x_max = max(x)

    bin_size = int((x_max - x_min) / n_bins)
    
    # print("Max", x_max)
    # print("Min", x_min)
    # print("Bin size", bin_size)

    hist = [0] * n_bins
    bin_list = [[] for _ in range(n_bins)]

    # print(hist)
    for t in x:
        for i in range(n_bins):
            # print("t", t, "i", i, "Min", x_min + (i * bin_size), "Max", x_min + ((i + 1) * bin_size))
            if x_min + (i * bin_size) <= t < x_min + ((i + 1) * bin_size):
                hist[i] += 1
                bin_list[i].append(t)
                break
    
    # print(hist)
    # print(bin_list)
    
    return hist, bin_list

# Testing
# data = [12, 45, 23, 67, 89, 34, 22, 90, 76, 58]
# print("Histogram:", histogram(data))
````

*Exercise 5:* Write a function the applies a booling function (that returns true/false) to every element in data, and return a list of indices of elements where the result was true. Use this function to find the indices of entries greater than 0.5. 

````python
def where(mylist, myfunc):
    out = []
    
    def myfunc(a):
        if a > 0.5:
            return True
        else:
            return False

    cnt = 0
    for i in mylist:
        if myfunc(i):
            out.append(cnt)
        cnt += 1

    return out
````

If `p<=function(test_x)`, then place `test_x` on the output. If not, repeat the process, drawing two new numbers. Repeat until you have the specified number of generated numbers, `N`. For this problem, it's OK to determine the `min` and `max` by numerically sampling the function.  

````python
import random

# A test function
def test_func(x, a=1, b=1):
    return abs(a * x + b)

def generate_function(func, x_min, x_max, N=1000):
    out = []
    i = 1
    while i <= N:
        test_x = random.randint(0, 100)
        p = random.randint(x_min, x_max)
        if p <= func(test_x):
            out.append(test_x)
            i += 1
    # print(out)
    return out

result = generate_function(test_func, 50, 100, 50)
print(result)

hist, edge_bins = histogram(result, 4)

print("Histogram")
print(hist)

print("Bin Content")
print(edge_bins)
````
